```python
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
rng = np.random.default_rng()

rand_numb = rng.normal(0, 10, 1000)

plt.hist(rand_numb, bins=100, edgecolor='Red')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("RN from Normal Distribution")

plt.savefig("normal_distribution_histogram.pdf", bbox_inches=="Tight", facecolor=="black")
plt.show()
```
