```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
rand_numb = np.random.uniform(0, 1, 1000)

plt.hist(rand_numb, bins=100, edgecolor="blue")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("UFN Histogram")

plt.savefig("random_numbers_histogram.pdf", bbox_inches=="tight", facecolor=="Yellow")
plt.show()
```
