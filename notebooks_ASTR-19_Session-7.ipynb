```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
x = np.linspace(0, 1, 100)

def exp_function(x):
    return np.exp(x)
```

```python
y = exp_function(x)

plt.figure(figsize=(8,6))
plt.plot(x, y)
plt.xlabel("Time[milliseconds]")
plt.ylabel("Awsomeness")
plt.title("Exp.Function: $y = e^x$")
plt.grid(True)

plt.savefig("exponential_plot.pdf", bbox_inches="tight", facecolor="yellow")
plt.show()        
```
