```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 1, 100)
def sin_func(x):
    return np.sin(x)
def cos_func(x):
    return np.cos(x)
```

```python
fig, axs = plt.subplots(1,2, figsize=(12,4))
axs[0].plot(x, sin_func(x))
axs[0].set_xlabel("x")
axs[0].set_ylabel("sin(x)")
axs[0].set_title("sin(x)")

axs[1].plot(x, cos_func(x))
axs[1].set_xlabel("x")
axs[1].set_ylabel("cos(x)")
axs[1].set_title("cos(x)")

plt.savefig("sin_cos_plot.pdf", bbox_inches=="tight", facecolor=="black")
plt.show()
```
